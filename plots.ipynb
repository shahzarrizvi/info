```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

# Uniforms


## Distributions

```python
fig, axs = plt.subplots(1, 3, figsize = (7, 3), sharey = True, sharex = True)

r = 0
C = stats.multivariate_normal.rvs(np.zeros(2), [[1, r],[r, 1]], size = 10000)
D = stats.norm.cdf(C)
X = D[:, 0]
Y = D[:, 1]
axs[0].scatter(X, Y, c = 'k', alpha = 0.05)
axs[0].set_aspect('equal')
axs[0].set_xlabel(r'$X$')
axs[0].set_ylabel(r'$Y$')
axs[0].set_title(r'$r = 0.00$')

r = 0.50
C = stats.multivariate_normal.rvs(np.zeros(2), [[1, r],[r, 1]], size = 10000)
D = stats.norm.cdf(C)
X = D[:, 0]
Y = D[:, 1]
axs[1].scatter(X, Y, c = 'k', alpha = 0.05)
axs[1].set_aspect('equal')
axs[1].set_xlabel(r'$X$')
axs[1].set_title(r'$r = 0.50$')

r = 0.95
C = stats.multivariate_normal.rvs(np.zeros(2), [[1, r],[r, 1]], size = 10000)
D = stats.norm.cdf(C)
X = D[:, 0]
Y = D[:, 1]
axs[2].scatter(X, Y, c = 'k', alpha = 0.05)
axs[2].set_aspect('equal')
axs[2].set_xlabel(r'$X$')
axs[2].set_title(r'$r = 0.95$')

fig.tight_layout()
```

## Varying $n$

```python
dir = 'uniforms/n/{}/'
ns = [100, 500, 1000, 2000, 5000]
s_p_vals = []
r_p_vals = []
for n in ns:
    r_p_vals += [np.load(dir.format(n) + 'r_p_val.npy')]
    s_p_vals += [np.load(dir.format(n) + 's_p_val.npy')]
```

```python
plt.figure(figsize = (3.5, 3))

plt.plot(ns, s_p_vals, c = 'dodgerblue', lw = 2, marker = 'o', 
         label = r'informational')
plt.plot(ns, r_p_vals, c = 'crimson', lw = 2, marker = 'o', 
         label = r'correlational')

plt.legend()
plt.xlim(100, 5000)
plt.ylim(0, 1)
plt.semilogx()
plt.xticks(ns, ns)
plt.axhline(0.05, c = 'gray', ls = '--', lw = 1)

plt.grid()
plt.gca().set_axisbelow(True);
```

# Betas

## Distributions

```python
fig, axs = plt.subplots(1, 3, figsize = (7, 3), sharey = True, sharex = True)

 = np.pi/4
R = np.array([[np.cos(), -np.sin()], [np.sin(), np.cos()]])

b = 1.0
a = 1
U = stats.uniform.rvs(scale = 2*np.pi, size = 10000)
V = np.vstack([a * np.cos(U), b * np.sin(U)])
D = (R @ V).T
X = D[:, 0]
Y = D[:, 1]
axs[0].scatter(X, Y, c = 'k', alpha = 0.05)
axs[0].set_aspect('equal')
axs[0].set_xlabel(r'$X$')
axs[0].set_ylabel(r'$Y$')
axs[0].set_title(r'$b = 1.0$')

b = 0.5
a = 1
U = stats.uniform.rvs(scale = 2*np.pi, size = 10000)
V = np.vstack([a * np.cos(U), b * np.sin(U)])
D = (R @ V).T
X = D[:, 0]
Y = D[:, 1]
axs[1].scatter(X, Y, c = 'k', alpha = 0.05)
axs[1].set_aspect('equal')
axs[1].set_xlabel(r'$X$')
axs[1].set_title(r'$b = 0.5$')

b = 0.1
a = 1
U = stats.uniform.rvs(scale = 2*np.pi, size = 10000)
V = np.vstack([a * np.cos(U), b * np.sin(U)])
D = (R @ V).T
X = D[:, 0]
Y = D[:, 1]
axs[2].scatter(X, Y, c = 'k', alpha = 0.05)
axs[2].set_aspect('equal')
axs[2].set_xlabel(r'$X$')
axs[2].set_title(r'$b = 0.1$')

fig.tight_layout()
```
